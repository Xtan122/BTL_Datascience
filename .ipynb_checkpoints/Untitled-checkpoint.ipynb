# Phân tích Thành tích Học sinh: Ảnh hưởng của Phong cách Học tập

## Câu hỏi
Phân tích ảnh hưởng của phong cách học tập đến thành tích học sinh

## Mục tiêu
- Khám phá mối quan hệ giữa `Preferred_Learning_Style` và `Final_Grade`.
- Đánh giá tác động của `Study_Hours_per_Week` đến `Exam_Score (%)`.
- Dự đoán điểm thi bằng hồi quy tuyến tính.

```python
# Nhập thư viện
import numpy as np
import pandas as pd
from datascience import *
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('fivethirtyeight')
```

```python
# Nhập dữ liệu
data = Table.read_table('student_performance_large_dataset.csv')

# Kiểm tra 5 hàng đầu tiên
data.show(5)

# Kiểm tra số hàng và cột
print(f"Số hàng: {data.num_rows}")
print(f"Số cột: {data.num_columns}")

# Kiểm tra các cột
print("Các cột trong dữ liệu:", data.labels)
```

## Khám phá Dữ liệu 

Trước khi phân tích sâu, chúng ta sẽ kiểm tra phân phối của các biến chính:
- **`Final_Grade`**: Điểm cuối kỳ (A, B, C, D, F).
- **`Preferred_Learning_Style`**: Phong cách học tập (Visual, Auditory, Kinesthetic, Reading/Writing).
- **`Study_Hours_per_Week`**: Số giờ học mỗi tuần.
- **`Exam_Score (%)`**: Điểm thi.

```python
# Đếm số lượng Final_Grade
grade_counts = data.group('Final_Grade')
grade_counts.show()
grade_counts.barh('Final_Grade', 'count')
plt.title('Phân phối của Final_Grade')  
plt.show()
```

```python
# Đếm số lượng Preferred_Learning_Style
style_counts = data.group('Preferred_Learning_Style')
style_counts.show()
style_counts.barh('Preferred_Learning_Style', 'count')
plt.title('Phân phối của Preferred_Learning_Style')  # Thêm tiêu đề ở đây
plt.show()
```

## Trực quan hóa Mối quan hệ

Chúng ta sẽ sử dụng các biểu đồ để khám phá:
1. Phân phối `Final_Grade` theo `Preferred_Learning_Style` thông qua biểu đồ cột xếp chồng.
2. Mối quan hệ giữa `Study_Hours_per_Week` và `Exam_Score (%)` thông qua biểu đồ phân tán.

```python
# Biểu đồ cột xếp chồng cho Final_Grade theo Preferred_Learning_Style
pivot_table = data.pivot('Preferred_Learning_Style', 'Final_Grade')
pivot_table.show()
pivot_table.bar('Final_Grade')
plt.title('Phân phối Final_Grade theo Preferred_Learning_Style')
plt.show()
```

## Hồi quy Tuyến tính

Dự đoán `Exam_Score (%)` dựa trên `Study_Hours_per_Week`


```python
# Hàm tính hồi quy tuyến tính
def correlation(x, y):
    """Tính hệ số tương quan giữa x và y"""
    x_standard = (x - np.mean(x)) / np.std(x)
    y_standard = (y - np.mean(y)) / np.std(y)
    return np.mean(x_standard * y_standard)

def linear_regression(x, y):
    """Tính độ dốc và chặn của đường hồi quy"""
    r = correlation(x, y)
    slope = r * np.std(y) / np.std(x)
    intercept = np.mean(y) - slope * np.mean(x)
    return slope, intercept

# Lấy dữ liệu
x = data.column('Study_Hours_per_Week')
y = data.column('Exam_Score (%)')

# Tính hồi quy
slope, intercept = linear_regression(x, y)
print(f"Độ dốc: {slope:.2f}")
print(f"Chặn: {intercept:.2f}")

# Vẽ biểu đồ phân tán và đường hồi quy
data.scatter('Study_Hours_per_Week', 'Exam_Score (%)')
plt.title('Hồi quy: Study_Hours_per_Week vs Exam_Score (%)')
x_range = np.array([min(x), max(x)])
y_pred = slope * x_range + intercept
plt.plot(x_range, y_pred, color='red', label='Đường hồi quy')
plt.legend()
plt.show()
```

## Suy luận Thống kê

Chúng ta sẽ kiểm tra xem phong cách học tập có ảnh hưởng đáng kể đến `Final_Grade` không bằng cách so sánh trung bình `Exam_Score (%)` giữa các nhóm `Preferred_Learning_Style`.

```python
# Tính trung bình Exam_Score (%) theo Preferred_Learning_Style
mean_scores = data.group('Preferred_Learning_Style', np.mean).select(['Preferred_Learning_Style', 'Exam_Score (%) mean'])
mean_scores.show()

# Vẽ biểu đồ cột cho trung bình Exam_Score
mean_scores.barh('Preferred_Learning_Style', 'Exam_Score (%) mean')
plt.title('Trung bình Exam_Score theo Preferred_Learning_Style')
plt.show()

# Hàm kiểm tra sự khác biệt trung bình (giả định đơn giản)
def difference_of_means(table, group_label, value_label):
    """Tính sự khác biệt trung bình giữa các nhóm"""
    means = table.group(group_label, np.mean).column(value_label + ' mean')
    return max(means) - min(means)

# Tính sự khác biệt trung bình
diff = difference_of_means(data, 'Preferred_Learning_Style', 'Exam_Score (%)')
print(f"Sự khác biệt trung bình lớn nhất: {diff:.2f}%")
```
